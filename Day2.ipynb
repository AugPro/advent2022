```python
score_per_shape = {"X":1, "Y":2, "Z":3}
score_per_outcome = {
    "X":{"A":3, "B":0, "C":6},
    "Y":{"A":6, "B":3, "C":0},
    "Z":{"A":0, "B":6, "C":3}
}

def compute_total_score(games:str):
    return sum(game_score(*game.split()) for game in games.strip().split("\n"))

def game_score(opponent:str, me:str)->int:
    return score_per_shape[me] + score_per_outcome[me][opponent]
    

    
assert game_score("A","Y") == 8
assert game_score("B","X") == 1
assert game_score("C","Z") == 6
assert compute_total_score(example) == 15
```

```python
score_per_shapev2 = {
    "X":{"A":3, "B":1, "C":2},
    "Y":{"A":1, "B":2, "C":3},
    "Z":{"A":2, "B":3, "C":1}
}
score_per_outcomev2 = {
    "X":0,
    "Y":3,
    "Z":6
}

def compute_total_scorev2(games:str):
    return sum(game_scorev2(*game.split()) for game in games.strip().split("\n"))

def game_scorev2(opponent:str, me:str)->int:
    return score_per_shapev2[me][opponent] + score_per_outcomev2[me]
    
```
